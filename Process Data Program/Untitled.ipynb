```python
## Variable text collects the article
## Variable entity collects the label(tag) 
text = []
entity = []

## Using for loop to run each row
for i in a:
    ## if the character length of column more than 100 then append to text
    if len(i) > 100:
        text.append(i[:-1])
    ## else. append to entity
    else:
        entity.append(i)
```

```python
## we arrange the entity,because entity has a lot separate tokens , we don't need,make entity format as Variable entity_all
## Row from left to right are: article id,start position of tag, end position of tag, tag , tag type
entity_all =[]
for i in entity:
    if i.split('\t')[0] != 'article_id' and i != '\n' and i != '--------------------\n':
        entity_all.append(i[:-1].split('\t'))
entity_all.append(['120'])
```

```python
entity_all
```

```python
## Then we reshape the entity_all,let same article id into smae list
## so we have 3 dimension in entity_all_2,first is same article id field,second is each row 
## third is each column for tag element like article id,start id....  
entity_all_2 = []
enti_ = []
for i in range(1,len(entity_all),1):
    if i == 1:
        enti_.append(entity_all[0])
    if enti_ == []:
        enti_.append(entity_all[i-1])
    if entity_all[i][0] == entity_all[i-1][0]:
        enti_.append(entity_all[i])
    else:
        entity_all_2.append(enti_)
        enti_ = []
```

```python
## Convert the column 1 and 2 to integer variable
for i in range(len(entity_all_2)):
    for j in range(len(entity_all_2[i])):
        entity_all_2[i][j][1],entity_all_2[i][j][2] = int(entity_all_2[i][j][1]),int(entity_all_2[i][j][2])
entity_all_2
```

```python

full_data = []
for enti_,text_ in zip(entity_all_2,text):
    test = []
    if len(enti_) == 1:
        print(enti_)
        sen_cut = ws([text_[:enti_[0][1]]])[0]
        for k in sen_cut:
            test.append((k,'O'))
        token = ws([enti_[0][3]])[0]
        for k,l in zip(token,range(len(token))):
            if l == 0:
                tk = 'B-' + enti_[0][4]
                test.append((k,tk))
            else: 
                tk = 'I-' + enti_[0][4]
                test.append((k,tk))
        sen_cut = ws([text_[enti_[0][2]:]])[0]
        for k in sen_cut:
            test.append((k,'O'))
    else:
        for i,j in zip(enti_,range(len(enti_))):
            if j == 0:
                sen_cut = ws([text_[:i[1]]])[0]
                for k in sen_cut:
                    test.append((k,'O'))
                token = ws([i[3]])[0]
                for k,l in zip(token,range(len(token))):
                    if l == 0:
                        tk = 'B-' + i[4]
                        test.append((k,tk))
                    else: 
                        tk = 'I-' + i[4]
                        test.append((k,tk))
                sen_cut = ws([text_[enti_[j][2]:enti_[j+1][1]]])[0]
                for k in sen_cut:
                    test.append((k,'O'))
            elif j == len(enti_)-1:
                token = ws([i[3]])[0]
                for k,l in zip(token,range(len(token))):
                    if l == 0:
                        tk = 'B-' + i[4]
                        test.append((k,tk))
                    else: 
                        tk = 'I-' + i[4]
                        test.append((k,tk))
                sen_cut = ws([text_[i[2]:]])[0]
                for k in sen_cut:
                    test.append((k,'O'))
            else:
                token = ws([i[3]])[0]
                for k,l in zip(token,range(len(token))):
                    if l == 0:
                        tk = 'B-' + i[4]
                        test.append((k,tk))
                    else: 
                        tk = 'I-' + i[4]
                        test.append((k,tk))
                sen_cut = ws([text_[enti_[j][2]:enti_[j+1][1]]])[0]
                for k in sen_cut:
                    test.append((k,'O'))
    break            
    full_data.append(test)
```

```python
with open("./dev3.txt","w+",encoding="utf-8") as f: 
    for data in full_data:
        for i in data:
            f.write(i[0])
            f.write('\t')
            f.write(i[1])
            if i[0] == '？' or i[0] == '。':
                f.write('\n')
            f.write('\n')
```
